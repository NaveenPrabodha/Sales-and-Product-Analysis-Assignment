```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
df = pd.read_csv('drive/MyDrive/Dataset/SalesKaggle3.csv')
```

```python
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop duplicates
df = df.drop_duplicates()

# Handle inconsistencies if necessary
# For example, convert date column to datetime format
# No date column is present in the provided dataset, so we skip this step.

# Display the cleaned data
print(df.head())
```

```python
# Since there is no date column, let's calculate total revenue by year based on the available columns
total_revenue_by_year = df.groupby('ReleaseYear')['PriceReg'].sum().reset_index()

# Display the result
print(total_revenue_by_year)
```

```python
# Calculate total sales by product
total_sales_by_product = df.groupby('SKU_number')['SoldCount'].sum().reset_index()

# Find the best-selling product
best_selling_product = total_sales_by_product.loc[total_sales_by_product['SoldCount'].idxmax()]

# Display the result
print(best_selling_product)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart for total revenue by year
plt.figure(figsize=(10, 6))
plt.bar(total_revenue_by_year['ReleaseYear'], total_revenue_by_year['PriceReg'])
plt.title('Total Revenue by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Total Revenue')
plt.show()

# Create a bar chart for the best-selling product
plt.figure(figsize=(10, 6))
plt.bar(best_selling_product['SKU_number'], best_selling_product['SoldCount'])
plt.title('Best-Selling Product')
plt.xlabel('SKU Number')
plt.ylabel('Total Sales')
plt.show()
```
